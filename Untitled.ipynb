```python
import pandas as pd
import requests
import bs4
url="https://www.timeanddate.com/weather/pakistan/karachi/ext"
heading=["Date","Temperature","Weather","Feels_like","Humidity","wind"]
req=requests.get(url)
soup=bs4.BeautifulSoup(req.content,"html.parser")
table=soup.find_all("tbody")[0].find_all("tr")
date=[date.find('th').text for date in table]
temperature=[temp.find_all('td')[1].text.replace("\xa0","") for temp in table]
weather=[weather.find_all('td')[2].text for weather in table]
feels_like=[feel.find_all('td')[3].text.replace("\xa0","") for feel in table]
humidity=[humidity.find_all('td')[6].text for humidity in table]
wind=[wind.find_all('td')[4].text for wind in table]
   

df=pd.DataFrame(
{
   
 heading[0]:date,
    heading[1]:temperature,
    heading[2]:weather,
    heading[3]:feels_like,
    heading[4]:wind,
    heading[5]:humidity,

}
)
df
```
